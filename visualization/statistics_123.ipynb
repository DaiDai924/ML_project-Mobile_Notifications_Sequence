```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

notification_df = pd.read_csv('../data/notification_sequence.csv', sep = ',')
notification_df = pd.DataFrame(notification_df)

# drop '?'
notification_df = notification_df.replace('?', pd.NaT)    
notification_df = notification_df.dropna(axis=0)
```

```python
y = notification_df.iloc[:, -6:]
y_123 = np.array(y['tol_123'])
```

```python
tol_1 = {'s': 0, 'c': 0, 'e': 0, 'w': 0, 'y': 0, 'n': 0}
tol_2 = {'s': 0, 'c': 0, 'e': 0, 'w': 0, 'y': 0, 'n': 0}
tol_3 = {'s': 0, 'c': 0, 'e': 0, 'w': 0, 'y': 0, 'n': 0}

for order in y_123:
    tol_1[order[0]] += 1
    for i in range(1, 3):
        tol_2[order[i]] += 1
    for i in range(3, 6):
        tol_3[order[i]] += 1
```

```python
print(tol_1)
print(tol_2)
print(tol_3)
```

```python
import matplotlib.pyplot as plt

labels = []
size = []
for key, value in tol_1.items():
    labels.append(key)
    size.append(value)
plt.pie(size, labels = labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('First position of the order')
plt.show()
```

```python
labels = []
size = []
for key, value in tol_2.items():
    labels.append(key)
    size.append(value)
plt.pie(size, labels = labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Second and Third positions of the order')
plt.show()
```

```python
labels = []
size = []
for key, value in tol_3.items():
    labels.append(key)
    size.append(value)
plt.pie(size, labels = labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Other positions of the order')
plt.show()
```
